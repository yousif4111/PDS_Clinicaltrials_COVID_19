## Data cleaning (country).  
The Locations (dataclean$Locations) in the dataset do not have a common structure.  
Few common structure such as  
1) Facility address. (Washington University School of Medicine, Saint Louis, Missouri, United States)  
2) inclusion of non-english (GH PitiÃ©-SalpÃªtriÃ¨re / Service d'Accueil des Urgences, Paris, Ile-de-France, France)  

But overall, we notice that the country names is very commonly available and decided to use it as key.  
So we use check a list of known country name and if it detected, it will be assigned to the country.

Result:  
We managed to assign 3653 out of 3669 (99.5\%)


```r
# Create function to map Locations to known location from countrylist.
countrylist <- read.csv("https://raw.githubusercontent.com/datasets/country-list/master/data.csv")
tomaploc <- function (loc) {
    # default to undefined
    country <- 'Undefined'
    
    # loop the location name through known countrylist. 
    for (j in countrylist$Name) { 
        if ( grepl(j,loc,ignore.case=TRUE) ) {
          country <- j ; break }
        }
    return (country)  }

# Create new column (MapCountry) with the locatin mapping.
dataclean$MapCountry = map(dataclean$Locations,tomaploc)

head(dataclean)

# Check the mapping 
# Number of valid Location (None empty)
print(paste("Valid Locations count =",sum(dataclean$Locations != "") ))

# Number of MapCountry (None "Undefined")
print(paste("Valid MapCountry count =",sum(dataclean$MapCountry != "Undefined") ))
```

## Data cleaning (drug).  
The Intervention (dataclean$Intervention) column have a structure of  
1) using pipe separator ("|").  
2) all the drug is prefix with "Drug :".  
  
Create a new column (drugonly) to record the list of drug.
But since the dataframe column cannot accept a vector of drug, populate the column with drug name separated by pipe.  

Example : "Drug: Duvelisib|Procedure: Peripheral blood draw|Drug: Placebo"
convert to Duvelisib|Placebo  
  



```r
# extract drug from the Intervention column
extractdrug <- function(txt) {
  # Split the intervention by "|"
  splittxt = strsplit(txt,"|",fixed=TRUE)
  drug = c()
    
  # For each item check whether it is drug, if yes, collect the name
  for (i in splittxt[[1]]) { 
    if (substr(i,1,5) == "Drug:") {
      x = substr(i,7,nchar(i))
      if (substr(x,1,1) == "`") {x = substr(x,2,nchar(x))}
      if (substr(x,nchar(x)-1,nchar(x)) == "`") { x = substr(x,1,nchar(x)-1) }
      drug = c(drug,x)}   
    }
    
  # return the result as pipe separator
  return (paste(drug,collapse="|"))
  }                
                
dataclean$Drugonly = map(as.character(dataclean$Intervention),extractdrug)
head(dataclean)
```

```r
# Function to extract all the drug and count for the drugonly column
# drugs example is Duvelisib|Placebo

countdrug <- function(drugs) {
    druglist = list()
    for (i in drugs) {
        if (i != "") {
            a = strsplit(i,"|",fixed=TRUE)
            a = unique(a[[1]])
            for (j in a) {
                if (j %in% names(druglist)) { druglist[j] = druglist[j][[1]] + 1}
                else (druglist[a] = 1)
                }
            
            }        
        }
    return (druglist)
    }

# print(countdrug(dataclean$Drugonly))

# dataclean$Drugonly[1:10]
```
