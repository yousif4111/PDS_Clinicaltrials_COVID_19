```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('MyCleanData.csv')
```

```python
data.columns
```

```python
data.drop(['Unnamed: 0','Drugonly','Interventions'], inplace=True,axis=1)
```

```python
data
```

```python
data.isnull().sum()
```

```python
dac=data.dropna()
```

```python
data['MapCountry'].value_counts()
```

```python
cn=dac['MapCountry'].unique()
cn
```

```python
cn
```

```python
coun=pd.DataFrame(cn)
```

```python
coun['country']=cn
```

```python
coun.drop(0,inplace=True,axis=1)
```

```python
coun
```

```python
coun['country']=sorted(coun.country)
```

```python
coun['country'].value_counts()
```

```python
coun
```

```python
dac
```

```python
stu=dac[['Title','MapCountry']]
```

```python
stu['Treatment Studies']=stu.Title.str.contains("Plasma|Treatment|Treat")
stu['Prevention Studies']=stu.Title.str.contains("Prevent|Vaccine|prevent")
stu['Vaccine Studies']=stu.Title.str.contains('Vaccine')
stu['Diagnostic Studies']=stu.Title.str.contains("Diagnostic|test|Test")
```

```python
stu
```

```python
coun["Total Studies"]=np.nan
```

## For loop for counting total study

```python
for i in stu['MapCountry'].unique():
    for j in coun['country']:
        if i==j:
            coun.loc[coun['country']==j, 'Total Studies']=len(stu.loc[stu['MapCountry']==i])
```

```python
coun
```

```python
coun["Treatment Studies"]=np.nan
```

## Second for loop for counting

```python
for i in stu['MapCountry'].unique():
    for j in coun['country']:
        if i==j:
            coun.loc[coun['country']==j, 'Treatment Studies']=stu.loc[stu['MapCountry']==i, 'Treatment Studies'].sum()
```


## next

```python
coun["Prevention Studies"]=np.nan
```

```python
for i in stu['MapCountry'].unique():
    for j in coun['country']:
        if i==j:
            coun.loc[coun['country']==j, 'Prevention Studies']=stu.loc[stu['MapCountry']==i, 'Prevention Studies'].sum()
```


## next

```python
coun["Vaccine Studies"]=np.nan
```

```python
for i in stu['MapCountry'].unique():
    for j in coun['country']:
        if i==j:
            coun.loc[coun['country']==j, 'Vaccine Studies']=stu.loc[stu['MapCountry']==i, 'Vaccine Studies'].sum()
```


## next

```python
coun["Diagnostic Studies"]=np.nan
```

```python
for i in stu['MapCountry'].unique():
    for j in coun['country']:
        if i==j:
            coun.loc[coun['country']==j, 'Diagnostic Studies']=stu.loc[stu['MapCountry']==i, 'Diagnostic Studies'].sum()
```
